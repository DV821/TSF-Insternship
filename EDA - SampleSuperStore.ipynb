```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import data

data = pd.read_csv('SampleSuperstore.csv')
data.head()
```

```python
# Check for Null Values

data.isna().sum()
```

```python
# Summary of Dataset 

data.info()
```

```python
# Summary of numerical columns

data.describe()
```

```python
# Number of unique values in each column

data.nunique()
```

```python
# Rough insight of data

sns.pairplot(data, diag_kind='hist')
plt.show()
```

```python
# Correlation between columns

sns.heatmap(data.corr(), cmap='coolwarm',annot=True)
plt.show()
```

```python
print('Total Profit :',np.sum(data.Profit))
print('Total Sales :',np.sum(data.Sales))
```

```python
# Cities with total profits and sales

data.groupby('City')[['Profit','Sales']].sum().sort_values(by='Profit', ascending=False).head(10)
```

```python
# States with total profits and sales

data.groupby('State')[['Profit','Sales']].sum().sort_values(by='Profit', ascending=False).head(10)
```

### Total Profit is about 1/8th part of Sales.

### New York city, Los Angeles, Seattle, San Francisco and Detroit are the Top 5 profit making cities.

### California, New York, Washington, MIchigan, Virginia are the Top 5 profit making states.

### West region has the maximum profit.

### Standard class shipping mode yields the greatest profit.

### Most profitable deliveries are of Consumer segment.

### There is a large profit in Technological sales, especially copiers.

### Copiers, Phones and Accesories make up about 46% of total profit.
